```python
import pandas as pd
from google.colab.data_table import DataTable
DataTable.max_columns = 60
# Load the dataset
url="https://raw.githubusercontent.com/GouravSinghal22/Intro-to-AI-and-ML/main/hotel_bookings.csv"
data = pd.read_csv(url)
data
```

```python
# Display basic information about the dataset
print("Data Information:")
print(data.info())
```

```python
# Handling missing values
print("\nHandling Missing Values:")
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Fill missing values for specific columns
data['agent'].fillna(0, inplace=True)  # Fill missing 'agent' values with 0
data['company'].fillna(0, inplace=True)  # Fill missing 'company' values with 0

data
```

```python
# Drop rows with missing values in 'country'
data.dropna(subset=['country'], inplace=True)
data
```

```python
# Encoding categorical variables
print("\nEncoding Categorical Variables:")
# One-hot encode 'hotel', 'meal', 'market_segment', 'distribution_channel', 'deposit_type', and 'customer_type'
data = pd.get_dummies(data, columns=['hotel', 'meal', 'market_segment', 'distribution_channel', 'deposit_type', 'customer_type'])

data
```

```python
# Convert 'reservation_status_date' to datetime format
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])

# Display the few rows of the preprocessed data
print("\nPreprocessed Data:")

data

# Save the preprocessed data to a new CSV file if needed
# data.to_csv('preprocessed_hotel_bookings.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from google.colab.data_table import DataTable
DataTable.max_columns = 60

# Load the dataset
url = "https://raw.githubusercontent.com/GouravSinghal22/Intro-to-AI-and-ML/main/hotel_bookings.csv"
data = pd.read_csv(url)

# Categorical columns you want to label encode
categorical_cols = ['hotel', 'meal', 'market_segment', 'distribution_channel', 'deposit_type', 'customer_type']

# Create a label encoder for each categorical column and transform the data
label_encoders = {}
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    data[col] = label_encoders[col].fit_transform(data[col])

# Print the encoded dataset
data
```
